```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import librosa as lb
from tqdm import tqdm
import IPython.display as ipd
```

```python
df = pd.read_csv('metadata/UrbanSound8K.csv')
```

```python
df.head()
```

```python
def path(file_name):
    excerpt = df[df['slice_file_name']==file_name]
    path_name = 'audio/'+'fold'+str(excerpt['fold'].iloc[0])+'/'+file_name
    class_name = excerpt['classID'].values[0]
    # print(path_name)
    
    return path_name,class_name
    
```

```python
dataset = np.zeros((df.shape[0],2),dtype=object)
```

```python
dataset.shape
```

# Mfcc : Melfrequency cepstrum coefficients

1. Find the melspectogram of the audio signal
2. Taking the log of the melspectogram features
3. Apply the DCT(discrete cosine transform) to it

But Librosa has provides the mfcc feature in their library

```python
for x in tqdm(range(0,df.shape[0])):
    
    file_name = df.iloc[x]['slice_file_name']
    path_name,class_id = path(file_name)
    
    #applying the mfcc features on the file name
    
    data, sr = lb.load(path=path_name,sr=22050,res_type='kaiser_fast')            # 'kaiser fast computes the parameter faster'
    
    mfcc = lb.feature.mfcc(y=data,sr=sr,n_mfcc=40)                   
    mfcc = np.mean(mfcc.T,axis=0)
    
    dataset[x,0] = mfcc
    dataset[x,1] = class_id
```

```python
ls=[]
for x in range(dataset.shape[0]):
    ls.append(dataset[x,0].shape)
```

```python
print('Max:  ',max(ls) )
print('Min:  ',min(ls)) 
```

```python
np.save('mfcc_dataset',dataset,allow_pickle=True)
```
